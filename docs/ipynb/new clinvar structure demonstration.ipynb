```python
import myvariant
```

```python
mv = myvariant.MyVariantInfo()
```

```python
# query one rcv accession number
mv.query('clinvar.rcv.accession:RCV000198299')
```

```python
# query multiple rcv accession number
mv.query('clinvar.rcv.accession:RCV000198299 OR clinvar.rcv.accession:RCV000059118')
```

```python
# query a variant id
mv.query('clinvar.variant_id:18')
```

```python
# query for a specific field, e.g. clinical siginificance EQUALS TO pathogenic
mv.query('clinvar.rcv.clinical_significance:Pathogenic')
```

```python
# query for multiple fields, e.g. clinical significance EQUALS TO pathogenic AND chrom EQUALS 3
mv.query('clinvar.rcv.clinical_significance:Pathogenic AND clinvar.chrom:3')
```

```python
# query for an existing field, e.g. dbvar id
mv.query('_exists_:clinvar.dbvar')
```
